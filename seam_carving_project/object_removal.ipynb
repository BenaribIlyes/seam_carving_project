```python
# Import necessary libraries


import cv2 # type: ignore
import numpy as np
import matplotlib.pyplot as plt



#Â To use the Numba JIT compiler
from numba import njit

# To use joblib for parallel processing
from joblib import Parallel, delayed

# Load the image
image_ref = cv2.imread('desert.jpg')

#   Load the mask
mask = cv2.imread("mask_desert.png", cv2.IMREAD_GRAYSCALE)

from class_SeamCarver import SeamCarver

print("Original Image Shape:", image_ref.shape)
```

```python
sc = SeamCarver(image_ref)

image_without_object, image_without_object_scaled_up= sc.remove_object(image_ref, mask, method='l1',orientation='vertical')


plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image_ref, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(image_without_object.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Object removed')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(image_without_object_scaled_up.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title('Final image')
plt.axis('off')

plt.tight_layout()
plt.show()

print("Original Image Shape:", image_ref.shape)
print("Object removed:", image_without_object.shape)
print("Final image:", image_without_object_scaled_up.shape)
```
